```python
def check_number(number):
    
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"
print (check_number (10))
```

```python
def area_of_square(side_length):
    return side_length * side_length
print(area_of_square(4))
```

```python
def calculate_grade():
    score = int (input("insert number here"))
    
    if 90 <= score <= 100:
        return "A"
    elif 80 <= score < 90:
        return "B"
    elif 70 <= score < 80:
        return "C"
    elif 60 <= score < 70:
        return "D"
    else:
        return "F"
    
print (calculate_grade())

```
