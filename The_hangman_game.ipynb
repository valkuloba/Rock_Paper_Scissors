```python
import random

# List of words for the hangman game
words = ["debug", "programming", "database", "function", "hardware", "software"]

# Randomly choose a word
secret_word = random.choice(words)

# Convert the word into a list of dashes
display_word = ["-" for _ in secret_word]

# Set to store guessed letters
guessed_letters = set()

# Maximum number of failed attempts
max_attempts = 6
failed_attempts = 0

print("Welcome to The Hangman Game!")
print("Guess the word:", " ".join(display_word))

# Game loop
while failed_attempts < max_attempts and "-" in display_word:
    try:
        # Ask user for a letter
        guess = input("\nEnter a letter: ").lower()

        # Validate input
        if len(guess) != 1 or not guess.isalpha():
            raise ValueError("Please enter only ONE letter.")

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        # Add guess to guessed letters
        guessed_letters.add(guess)

        # Check if guess is in the word
        if guess in secret_word:
            print("Good guess!")
            for index, letter in enumerate(secret_word):
                if letter == guess:
                    display_word[index] = guess
        else:
            failed_attempts += 1
            print(f"Wrong guess! Attempts left: {max_attempts - failed_attempts}")

        # Show current word state
        print("Word:", " ".join(display_word))

    except ValueError as e:
        print(e)

# Game result
print("\nGame Over!")
if "-" not in display_word:
    print("Congratulations! You WON!")
else:
    print("You Lost!")
    print("The correct word was:", secret_word)
```
