```python
# Import the random library so the computer can make random choices
import random

# Possible choices in the game
choices = ["rock", "paper", "scissors"]

# Initialize scores
user_score = 0
computer_score = 0

# Maximum points needed to win (best of 5 â†’ first to 3)
winning_score = 3

# Round counter
round_number = 1

# Loop continues until either the user or computer reaches 3 points
while user_score < winning_score and computer_score < winning_score:
    print(f"\nRound {round_number}")

    # Computer randomly chooses rock, paper, or scissors
    computer_choice = random.choice(choices)

    # User inputs their choice
    user_choice = input("Choose rock, paper, or scissors: ").lower()

    # Validate user input
    if user_choice not in choices:
        print("Invalid choice. Please choose rock, paper, or scissors.")
        continue

    print("Computer chose:", computer_choice)

    # Check for a tie
    if user_choice == computer_choice:
        print("It's a tie!")

    # Check winning conditions
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "scissors" and computer_choice == "paper") or
        (user_choice == "paper" and computer_choice == "rock")
    ):
        print("You win this round!")
        user_score += 1

    else:
        print("Computer wins this round!")
        computer_score += 1

    # Display current scores
    print(f"Score â†’ You: {user_score} | Computer: {computer_score}")

    round_number += 1

# Declare the overall winner
print("\nðŸŽ‰ Game Over!")
if user_score == winning_score:
    print("Congratulations! You won the game!")
else:
    print("Computer wins the game. Better luck next time!")
```
